## 确认是否可以使用`CUDA`来加载`torch`


```python
import torch
print(torch.cuda.is_available())
```

如果打印`False`，说明`torch`没有使用`CUDA`,此时所有计算将在`CPU`上进行,模型训练速度会非常慢,需要安装`CUDA`.

用`nvidia-smi`查看`CUDA`版本


```python
!nvidia-smi
```

然后去[pytorch官网](https://pytorch.org/)选择对应的版本并下载,下载前要先卸载目前安装的`torch`和`CUDA`


```python
# 先卸载现有 torch
!uv pip uninstall torch torchvision torchaudio

# 安装新版本 torch,根据网站上提示的命令安装
!uv install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu126

# 确认 torch 是否使用 CUDA
import torch
print(torch.cuda.is_available())

```
